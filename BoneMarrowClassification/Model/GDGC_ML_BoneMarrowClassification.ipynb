```python
# prompt: create a dataframe of bonemarrow images uploaded

import pandas as pd
import os

# Assuming images are in a directory named 'bonemarrow_images' in your current working directory
image_dir = '/content/bone_marrow_cell_dataset/ART/0001-1000'

# Create an empty list to store image data
image_data = []

# Iterate through the images in the directory
for filename in os.listdir(image_dir):
  if filename.endswith(('.jpg', '.jpeg', '.png')):  # Adjust file extensions if needed
    image_path = os.path.join(image_dir, filename)
    image_data.append({'filename': filename, 'path': image_path})

# Create a Pandas DataFrame from the image data
df2 = pd.DataFrame(image_data)

# Print the DataFrame
df2
```

```python
# prompt: create a new column as type and assign values ABE

df2['type'] = 'ART'
df2
```

```python
# prompt: concatenate df and df2

import pandas as pd
df_concat = pd.concat([df, df2], ignore_index=True)
df_concat
```

```python
# 1. CNN model to take in the bone marrow image and classifiy the type as ABE or ART

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# Define the CNN model
def create_cnn_model():
    model = keras.Sequential([
        layers.Conv2D(32, (3, 3), activation='relu', input_shape=(224, 224, 3)),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(64, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(128, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),
        layers.Flatten(),
        layers.Dense(128, activation='relu'),
        layers.Dense(2, activation='softmax')
    ])
    return model

# Create the model
model = create_cnn_model()

# Compile the model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Data preprocessing
import numpy as np
from sklearn.model_selection import train_test_split

def load_and_preprocess_image(image_path):
  # Load the image using the library
  # Preprocess the image (resize, normalize, etc.)
  # Return the preprocessed image as a NumPy array
  img = tf.keras.preprocessing.image.load_img(image_path, target_size=(224, 224))
  img = tf.keras.preprocessing.image.img_to_array(img)
  img = img / 255.0
  return img

images = []
labels = []
for index, row in df_concat.iterrows():
    try:
        image = load_and_preprocess_image(row['path'])
        images.append(image)
        labels.append(1 if row['type'] == 'ART' else 0)
    except Exception as e:
        print(f"Error processing image {row['path']}: {e}")

# Convert lists to numpy arrays
images = np.array(images)
labels = np.array(labels)

# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(images, labels, test_size=0.2, random_state=42)

# Train the model
model.fit(x_train, y_train, epochs=10, validation_data=(x_test, y_test)) # Adjust epochs as needed


# Evaluate the model
loss, accuracy = model.evaluate(x_test, y_test, verbose=0)
print(f"Test Loss: {loss:.4f}")
print(f"Test Accuracy: {accuracy:.4f}")
```

```python

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix

# Make predictions on the test set
y_pred_probabilities = model.predict(x_test)
y_pred = np.argmax(y_pred_probabilities, axis=1)  # Get predicted class labels

# Calculate metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
auc_roc = roc_auc_score(y_test, y_pred_probabilities[:, 1])  # Assuming binary classification
conf_matrix = confusion_matrix(y_test, y_pred)

# Create a DataFrame for the results
results_df = pd.DataFrame({
    'Metric': ['Accuracy', 'Precision', 'Recall', 'F1-score', 'AUC-ROC'],
    'Value': [accuracy, precision, recall, f1, auc_roc]
})


# Print the results table
print(results_df)
```

```python
# 2.  Bidirectional GRU + attention model to take in the bone marrow image and classifiy the type as ABE or ART

from tensorflow.keras import layers

def create_bidirectional_gru_attention_model(cnn_model):
    # Freeze the pre-trained CNN layers
    for layer in cnn_model.layers:
        layer.trainable = False

    input_layer = keras.Input(shape=(224, 224, 3))
    cnn_features = cnn_model(input_layer)  # Using the pre-trained CNN as feature extractor

    # Reshape CNN features to prepare for recurrent layer
    reshaped_features = layers.Reshape((1, -1))(cnn_features)

    # Bidirectional GRU
    gru_layer = layers.Bidirectional(layers.GRU(64, return_sequences=True))(reshaped_features)  # Adjust units as needed

    # Attention mechanism
    attention_probs = layers.Dense(128, activation='softmax')(gru_layer)  # Calculate attention weights
    context_vector = layers.Multiply()([attention_probs, gru_layer])  # Apply attention weights
    context_vector = layers.Lambda(lambda x: tf.reduce_sum(x, axis=1))(context_vector) # Context vector

    # Output layer
    output_layer = layers.Dense(2, activation='softmax')(context_vector)

    model = keras.Model(inputs=input_layer, outputs=output_layer)
    return model


combined_model = create_bidirectional_gru_attention_model(model)


combined_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

combined_model.fit(x_train, y_train, epochs=10, validation_data=(x_test, y_test)) # Adjust epochs as needed

# Evaluate the model
loss, accuracy = model.evaluate(x_test, y_test, verbose=0)
print(f"Test Loss: {loss:.4f}")
print(f"Test Accuracy: {accuracy:.4f}")
```
